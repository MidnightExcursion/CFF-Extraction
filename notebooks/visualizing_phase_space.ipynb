# Import Matplotlib and Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

Read in the data:

```python
jlab_pandas_df = read_csv_file_with_pandas('../data/jlab_kinematics.csv')
```

Plot the data using our custom-designed thing:

```python
# (1): Figure instance:
figure = plt.figure()

# (2): Add an Axes Object:
axes_object = figure.add_subplot(111, projection = '3d')

# (3): Obtain the data:
values_of_x_B = jlab_pandas_df['x_b']
values_of_Q_squared = jlab_pandas_df['QQ']
values_of_t = jlab_pandas_df['t']

# (4): Customize the Axes Object:
customizer = PlotCustomizer(
    axes_object,
    title = r"Kinematic Phase Space for JLab Data",
    xlabel = r"$x_{B}$",
    ylabel = r"$Q^{2}$",
    zlabel = r"$t$",
    grid = True)

# (5): Add data to the Axes Object:
customizer.add_3d_scatter_plot(
    values_of_x_B,
    values_of_Q_squared,
    values_of_t,
    color = 'blue')
```
