# Import Matplotlib and Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

Read in the data:

```python
jlab_pandas_df = read_csv_file_with_pandas('../data/jlab_kinematics.csv')
```

Plot the data using our custom-designed thing:

```python
# (1): Figure instance:
figure = plt.figure(figsize = (18, 6))

# (2): Add an Axes Object:
axes_1 = figure.add_subplot(131)
axes_2 = figure.add_subplot(132)
axes_3 = figure.add_subplot(133, projection = '3d')

# (3): Obtain the data:
values_of_x_B = jlab_pandas_df['x_b']
values_of_Q_squared = jlab_pandas_df['QQ']
values_of_t = jlab_pandas_df['t']

# (4): Customize the Axes Object:
customizer_1 = PlotCustomizer(
    axes_1,
    title = "",
    xlabel = r"$x_{B}$",
    ylabel = r"$Q^{2} [\text{GeV}^{2}]$",
    grid = True)

customizer_2 = PlotCustomizer(
    axes_2,
    title = "",
    xlabel = r"$x_{B}$",
    ylabel = r"$-t ^{2} [\text{GeV}^{2}]$",
    grid = True)

customizer_3 = PlotCustomizer(
    axes_3,
    title = r"Kinematic Phase Space for JLab Data",
    xlabel = r"$x_{B}$",
    ylabel = r"$Q^{2}$",
    zlabel = r"$t$",
    grid = True)

# (5): Add data to the Axes Object:
customizer_1.add_scatter_plot(
    values_of_x_B,
    values_of_Q_squared,
    color = 'blue')

customizer_2.add_scatter_plot(
    values_of_x_B,
    -1 * values_of_t,
    color = 'purple')

customizer_3.add_3d_scatter_plot(
    values_of_x_B,
    values_of_Q_squared,
    values_of_t,
    color = 'red')
```
